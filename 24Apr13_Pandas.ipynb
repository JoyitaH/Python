```python
import pandas as pd
```

```python
data = pd.read_csv("insurance.csv")
data.head()
```

```python
# Filtering
data['gender'] == 'female'
```

```python
# Lambda function => anonymous function
data['upd_children'] = data['children'].apply(lambda x : x+1)
```

```python
data.head()
```

```python
data['gender'].apply(lambda x : 1 if x == 'female' else 0)
```

```python
def age_group(x):
    if(x < 20):
        return '0-20'
    elif(x > 20 and x < 40):
        return '20-40'
    elif(x > 40 and x < 60):
        return '40-60'
    else:
        return '60+'
```

```python
data['age_group'] = data['age'].apply(age_group)
```

```python
data.head()
```

```python
# filter data with charges between 10000 and 20000
data[(data['charges'] > 10000) & (data['charges'] < 20000)]
```
