```python
# Quartiles  (can also be calculated using percentile values)
# q1 = 25 percentile
# q2 = 50 percentile
# q3 = 75 percentile

q1 = data["Population in thousands (2017)"].quantile(0.25)
q1
```

```python
q2 = data["Population in thousands (2017)"].quantile(0.50)
q2
```

```python
q3 = data["Population in thousands (2017)"].quantile(0.75)
q3
```

```python
# IQR = Q3 - Q1
iqr = q3 - q1
iqr
```

```python
# Outliers
# Upper Fence = Q3 + 1.5IQR
upperfence = q3 + (1.5*iqr)
upperfence
```

```python
# Lower Fence = Q1 - 1.5IQR
lowerfence = q1 - (1.5*iqr)
lowerfence
```
